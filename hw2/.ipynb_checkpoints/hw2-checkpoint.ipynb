```python
#Problem 3
image_path = "./sample1.jpg"
img = cv2.imread(image_path, 0)
plt.title("Original Image")
plt.imshow(img, cmap="gray")
```

```python
#Horizontal edge kernel
horiz = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
horiz = np.fliplr(np.flipud(horiz))
out_horiz = filter_lin(img, horiz)
plt.imshow(out_horiz, cmap="gray")
```

```python
#Vertical Edge Kernel
vert = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
vert = np.fliplr(np.flipud(vert))
out_vert = filter_lin(img, vert)
plt.imshow(out_vert, cmap="gray")
```

```python
#Horizontal filter to frequency domain
horiz = np.pad(horiz, ((0, img.shape[0]-3), (0, img.shape[1]-3)))
horiz_fft = np.fft.fft2(horiz)
horiz_shift = np.fft.fftshift(horiz_fft)
frequency_horiz = np.log(1+np.abs(horiz_shift))
plt.imshow(frequency_horiz, cmap="jet")
```

```python
#vertical filter to frequency domain
vert = vert.astype(np.float32)
vert = np.pad(vert, ((0, img.shape[0]-3), (0, img.shape[1]-3)))
freq_vert = np.fft.fft2(vert, )
freq_vertshif = np.fft.fftshift(freq_vert)
frequency_vert = np.log(1 + np.abs(freq_vertshif))
plt.imshow(frequency_vert, cmap="jet")
```

```python
#Sample1 Image to frequency domain
freq_img = np.fft.fft2(img)
freq_img_shift = np.fft.fftshift(freq_img)
frequency_img = np.log(1+np.abs(freq_img_shift))
plt.imshow(frequency_img, cmap="jet")
```

```python
#Horizontal Edge in frequency domain
horiz_after = freq_img * frequency_horiz
horiz_after_shift = np.fft.fftshift(horiz_after)
horiz_after_trans = 20*np.log(1+np.abs(horiz_after_shift))
plt.imshow(horiz_after_trans, cmap = "jet")
```

```python
#Vertical Edge in Frequency Domain
vert_after = freq_img * frequency_vert
vert_after_shift = np.fft.fftshift(vert_after)
vert_after_trans = np.log(1+np.abs(vert_after_shift))
plt.imshow(vert_after_trans, cmap = "jet")
```

```python
#Data from frequency to spatial domain (Horizontal Edge)
spat_horiz = np.fft.ifft2(horiz_after)
plt.imshow(np.abs(spat_horiz), cmap="gray")
```

```python
#Data from frequency to spatial domain (Vertical Edge)
spat_vert = np.fft.ifft2(vert_after)
plt.imshow(np.abs(spat_vert), cmap="gray")
```

```python
#Sum of horizontal and vertical components
img_sum = np.abs(spat_vert+spat_horiz)
plt.imshow(img_sum, cmap="gray")
```
