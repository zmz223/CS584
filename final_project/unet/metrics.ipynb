```python
def dice_metric(inputs, target):
    intersection = 2.0 * (target * inputs).sum()
    union = target.sum() + inputs.sum()
    if target.sum() == 0 and inputs.sum() == 0:
        return 1.0

    return intersection / union
```

```python
print(dice_metric(y_test[1][:, :, 0], pred[1][:, :, 0]))
```
